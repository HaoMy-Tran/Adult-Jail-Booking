## Formatting data in each column
Formatting Last Name, First Name, Middle Name, and Release Reason
Name of the prisons and Release Reason should be capitalized each word, rather than be uppercase.

```python
df['Last Name']=df['Last Name'].str.title()
df['First Name']=df['First Name'].str.title()
df['Middle Name']=df['Middle Name'].str.title()
df['Release Reason']=df['Release Reason'].str.title()
df.head()
```

Formatting Booking Date Time:

```python
df['Booking Date Time'] = pd.to_datetime(df['Booking Date Time'], format = '%m/%d/%Y %I:%M:%S %p')
df['Booking Date Time']
```

Format Release Date Time

```python
df['Release Date Time'] = pd.to_datetime(df['Release Date Time'], format = '%m/%d/%Y %I:%M:%S %p')
df['Release Date Time']
```

## Generating new columns

For the time series dataset, we can create more columns relating to time. In this case, I will generate 3 new columns, they are Day Name of Booking, Day Name of Release, length of In-jail Time:

```python
df['Booking Day Name'] = df['Booking Date Time'].dt.day_name()
df['Release Day Name'] = df['Release Date Time'].dt.day_name()
df['In-jail Time'] = df['Release Date Time'] - df['Booking Date Time']
df.head()
```

# Exploratory Data Analysis

After cleaning data, I will complete a Exploratory Data Analysis to learn more about this dataset. Let's import essential packages and begin with each column from left to right.

```python
from matplotlib import pyplot as plt
```

## Booking Date Time 
The first raised quesion is whether the number of people who commit criminal and number of criminal cases increase over time in King County?

```python
count_booking_date = df['Booking Date Time'].dt.date.value_counts().reset_index(name = 'Count')
count_booking_date
```

Sort values

```python
sorted_count_booking_date = count_booking_date.sort_values(by = 'index', ascending = True)
sorted_count_booking_date
```

Visualize the change in criminal trend by a line chart

```python
plt.figure(figsize=(12,4))
plt.plot(sorted_count_booking_date['index'],sorted_count_booking_date['Count'])
```

The Criminal trend overally decrease in the beginning of 2022. From the May 2022, it went down at reached the bottom. 
Let' look at the distribution of the Booking Date Time:

```python
plt.hist(df['Booking Date Time'].dt.hour, bins = 24, edgecolor = 'black')
plt.show()
```

The time spans with high frequency of adult criminal are from 5 am to 10 am, 6 pm to 7 pm, and from 9 pm 1 am. This distribution can raise to some question
The trend of law offendence is usually higher in the night and morning while lower in the day?
Why right before 8 pm, the rate suddenly losts momentum and falls, then recover again?
Is this distribution the different between in-week days and weekends? 
To anwer the last question, I plot two histogram charts for the days in week and the weekends:

```python
weekends = ['Sunday', 'Saturday']
weekdays = ['Monday','Tuesday', 'Wednesday','Thursday', 'Friday']
weekends_df = df.loc[df['Booking Day Name'].isin(weekends)]
weekdays_df = df.loc[df['Booking Day Name'].isin(weekdays)]
```

```python
plt.hist(weekends_df['Booking Date Time'].dt.hour, bins = 24, edgecolor = 'black')
plt.show()
```

```python
plt.hist(weekdays_df['Booking Date Time'].dt.hour, bins = 24, edgecolor = 'black')
plt.show()
```

Let's see the distribution of In-jail Time

From the to histogram charts, the criminal frequency dropped from the noon to 3 pm and kept stable until 7 am in the weekdays. Meanwhile, the frequency in the weekends kept decreasing after 9 am, reached the bottom at 1 pm and significantly increased after that. Therefore, we can conclude the distribution differs between weekdays and weekends

## In-jail Time

I will sum the time in jail by Book of Arrest Number because each offender has their own number of arrest

```python
Injail_time_pivot = df.pivot_table(index = ['Book of Arrest Number'], values = ['In-jail Time'], aggfunc = 'sum')
Injail_time_pivot
```

```python
Injail_time_pivot['In-jail Time'].describe()
```
